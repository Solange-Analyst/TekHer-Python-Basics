### Exercise 26:

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
labels = ['Start', 'Increase 1', 'Decrease 1', 'Increase 2', 'End']
values = [100, 20, -30, 15, 105]

cumulative_values = np.cumsum(values[:-1])

change_values = np.array(values[1:-1])
bar_heights = np.abs(change_values)
bar_bottoms = cumulative_values[:-1]
bar_bottoms = np.insert(bar_bottoms, 0, values[0]) 

bar_colors = ['green' if change > 0 else 'red' for change in change_values]
bar_colors.insert(0, 'purple') 
bar_colors.append('purple') 

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(labels[0], values[0], color=bar_colors[0])

for i in range(len(change_values)):
    if change_values[i] > 0:
        ax.bar(labels[i+1], bar_heights[i], bottom=bar_bottoms[i], color=bar_colors[i+1])
    else:
        ax.bar(labels[i+1], -bar_heights[i], bottom=bar_bottoms[i] + bar_heights[i], color=bar_colors[i+1])

end_value = values[-1]

ax.bar(labels[-1], end_value, color=bar_colors[-1])

for i in range(len(values) - 1):
    start_y = values[i] if i == 0 else cumulative_values[i-1] + change_values[i-1]
    end_y = cumulative_values[i]

    ax.hlines(start_y, i, i + 1, color='grey', linestyle='--')


ax.set_xlabel('Categories')
ax.set_ylabel('Value')
ax.set_title('Simple Waterfall Chart')
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# Exercise 49
num_attributes = 5

strengths = np.random.rand(num_attributes) * 10
angles = np.linspace(0, 2 * np.pi, num_attributes, endpoint=False).tolist()
strengths = np.concatenate((strengths, [strengths[0]])) 
angles.append(angles[0])

# Define the labels for each attribute
attribute_labels = [f'Attribute {i+1}' for i in range(num_attributes)]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot the data
ax.plot(angles, strengths, 'o-', linewidth=2)

# Fill the area under the plot
ax.fill(angles, strengths, alpha=0.25)
ax.set_thetagrids(np.degrees(angles[:-1]), attribute_labels)


ax.set_title('Radar Chart of Five Attributes')
ax.set_ylim(0, 10)
ax.set_yticks(np.linspace(0, 10, 5)) 

# Add a grid
ax.grid(True)

plt.show()
```

### Exercise 50:
Create a 3D scatter plot with random values for x, y, and z.


```python
# Exercise 50
np.random.seed(42) 
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
z = np.random.rand(100) * 10
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Set labels and title
ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')
ax.set_zlabel('Z Value')
ax.set_title('3D Scatter Plot of Random Data')

fig.colorbar(ax.scatter(x, y, z, c=z, cmap='viridis', marker='o'), label='Z Value')

plt.show()
```
